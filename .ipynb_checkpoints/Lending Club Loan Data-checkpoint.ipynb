# Data Cleaning

```python
## Number of duplicates
print('Number of duplicates:',data_loan.duplicated().sum())
```

```python
data_loan.isnull().sum()
```

```python
data_loan.dropna(axis = 1, how = 'all', inplace = True)
data_loan.shape
```

```python
(data_loan.isnull().sum()/data_loan.shape[0]*100).round(2).sort_values(ascending=False)
```

Most of the columns has 100% Null values and can be safely droped. There are some columns where NULL value 
rate is 2%, 33%, 65%, 93% and 97%. 93% and 97% columns can be dropped as well. It is now necessary to put a 
threshold of the rate. Looking into the criticality of the columns and NULL values rate, it is okay to consider 
the rate as 30%. So any Columns having NaN or NULL values greater than or equal to 30% can be safely dropped.

Therefore, Threshold Dropped Percentage Rate = 30%

```python
data_loan = data_loan.loc[:,data_loan.isnull().sum()/data_loan.shape[0]*100<30]
# Shape of the dataframe after removing columns
```

```python
#Basic information about the data
## Number of rows and columns
print('Number of Columns:',data_loan.shape[1])
print('Number of Rows:',data_loan.shape[0])
```

```python
#Finding number of unique values in each vcolumn
data_loan.nunique().sort_values()
```

There are several columns which are single valued.
They cannot contribute to our analysis in any way. So removing them.

```python
data_loan = data_loan.drop(['tax_liens', "delinq_amnt",'chargeoff_within_12_mths','acc_now_delinq','application_type', 'policy_code', 'collections_12_mths_ex_med', 'initial_list_status', 'pymnt_plan'], axis = 1)
data_loan.head()
```

```python
#Finding number of unique values in each vcolumn
data_loan.nunique().sort_values(ascending=False)
```

```python
# Removing irrelevant columns which are calculated after loan is approved thus have no relevance to the analysis
## The columns removed are customer behaviour variables and are calculated after thus they cannot be used as input for the analysis
data_loan = data_loan.drop(['delinq_2yrs','earliest_cr_line','inq_last_6mths','open_acc','pub_rec','revol_util','revol_bal','total_acc','out_prncp','out_prncp_inv','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','last_credit_pull_d'],axis=1)
# Removing desc,emp_title,desc as they have no significance to the analysis as they are random varaibles and hold infomration which might not have significance to the conclusions
data_loan = data_loan.drop(['title','emp_title','url'],axis=1)
# Removing zip_code as it is a masked data and cannot be used as input for the analysis
data_loan = data_loan.drop(['zip_code'],axis=1)
# Removing member_id as it is a duplicate index column and is not required for the analysis
data_loan = data_loan.drop(['id', 'member_id'],axis=1)
# Shape of the dataframe after removing columns
print(data_loan.shape)
```

```python
data_loan.columns
```

The goal of the analysis is to see who is likely to default and this can only be said in case of either fully paid or charged off loans.
We cannot make anything up for the current loans.
To exclude that data , removing the records with current loan status

```python
data_loan = data_loan[data_loan.loan_status != "Current"]
data_loan.loan_status.unique()
```

```python
print(data_loan.shape)
```

```python
(data_loan.isna().sum()/len(data_loan.index))*100
```

```python
data_loan['emp_length'].value_counts()
```

```python
mode_emp_length = data_loan['emp_length'].mode()[0]
mode_emp_length
```

```python
data_loan['emp_length'].fillna(mode_emp_length, inplace = True)
data_loan['emp_length'].isna().sum()
```

```python
data_loan['pub_rec_bankruptcies'].value_counts()
```

```python
mode_pub_rec_bankruptcies = data_loan['pub_rec_bankruptcies'].mode()[0]
mode_pub_rec_bankruptcies
```

```python
data_loan['pub_rec_bankruptcies'].fillna(mode_pub_rec_bankruptcies, inplace = True)
data_loan['pub_rec_bankruptcies'].isna().sum()
```

```python
data_loan['home_ownership'].value_counts()
```

As per Data Disctionary, Variable home_ownership has 4 valid values - RENT, OWN, MORTGAGE, OTHER. However, in
data, there are 3 records whose value is NONE. Since record is very minimum and doesn't add any significant 
value, so dropping rows of those 3 records.

```python
#replacing 'NONE' with 'OTHERS'
data_loan['home_ownership'].replace(to_replace = ['NONE'],value='OTHER',inplace = True)
```

# Standardizing the data

```python
data_loan.dtypes
```

```python
#stripping '%' value
data_loan['int_rate'] = data_loan['int_rate'].apply(lambda x:str(x).replace('%','')).astype('float').round(2)
```

```python
# Converting emp_length to integer values
# Converting emp_length as numerical data to create more effective statistical analysis as compared to nominal values
data_loan['emp_length'] = pd.to_numeric(data_loan['emp_length'].apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

```python
data_loan['issue_d'] = pd.to_datetime(data_loan['issue_d'],format='%b-%y')
data_loan['issue_year']=data_loan['issue_d'].dt.year
data_loan['issue_month']=data_loan['issue_d'].dt.month
```

```python
data_loan['term'] = data_loan['term'].astype('category')
data_loan['grade'] = data_loan['grade'].astype('category')
data_loan['sub_grade'] = data_loan['sub_grade'].astype('category')
data_loan['home_ownership'] = data_loan['home_ownership'].astype('category')
data_loan['verification_status'] = data_loan['verification_status'].astype('category')
data_loan['purpose'] = data_loan['purpose'].astype('category')
data_loan['addr_state'] = data_loan['addr_state'].astype('category')
```

```python
data_loan.head()
```
